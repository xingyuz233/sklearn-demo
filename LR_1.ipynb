1.导入相关类库

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import warnings
```

2.产生模拟数据集

```python
# 产生模拟数据集
N = 200
X, y = make_regression(n_samples=N, n_features=1, n_informative=1, bias=150.0, noise=30)
# 绘制原始数据集
mpl.colors.ListedColormap(list('rgbmyc'))
plt.figure(figsize=(15, 9), facecolor='w')
plt.title('Linear Regression')
plt.scatter(X, y, marker='o', s=50)
plt.show()
```

3.线性回归模型的构造

```python
# 测试集与训练集的拆分
X_train, X_test, y_train, y_test = train_test_split(X, y)
# 调用线性回归模型
linreg = LinearRegression()
# 训练模型
linreg.fit(X_train, y_train)
```

4.数据可视化

```python
# 数据导入
y_predict = linreg.coef_ * X + linreg.intercept_
# 模型图形展现
## 图形基本性质
cm = mpl.colors.ListedColormap(list('rgbmyc'))
plt.figure(figsize=(15, 9), facecolor='w')
## 绘图内容
plt.scatter(X, y, cmap=cm, edgecolors='none')
plt.plot(X, y_predict, 'r-')
## 图形标签
plt.title('Least Squares Regression')
plt.xlabel('Feature value (x)')
plt.ylabel('Feature value (y)')
plt.grid(True)
# 图形展示
plt.show()
```

5.模型评估

```python
print '线性模型的系数(w): {}'.format(linreg.coef_)
print '线性模型的常系数(b): {:.3f}'.format(linreg.intercept_)
print '训练集中R-squared得分: {:.3f}'.format(linreg.score(X_train, y_train))
print '测试集中R-squared得分: {:.3f}'.format(linreg.score(X_test, y_test))
```
