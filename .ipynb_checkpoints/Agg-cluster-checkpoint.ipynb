1.导入相关类库

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
import sklearn.datasets as ds
import warnings
```

2.产生模拟数据集

```python
N = 2000
centers1 = 4
centers2 = 2
# 圆形数据集
data1, y1 = ds.make_blobs(n_samples = N, n_features = 2, centers = centers)
# 月牙数据集
data2, y2 = ds.make_moons(n_samples = N)
```

3.层次聚类模型的构造与可视化

```python
def mdata_v(row,col,index,data,y,data_title):
    plt.subplot(row,col,index)
    plt.scatter(data[:,0], data[:,1], c=y, s=30, cmap=cm, edgecolors='none')
    
    x1_min, x2_min = np.min(data,axis=0)
    x1_max, x2_max = np.max(data,axis=0)
    x1_min, x1_max = expandBorder(x1_min, x1_max)
    x2_min, x2_max = expandBorder(x2_min, x2_max)
    plt.xlim(x1_min,x1_max)
    plt.ylim(x2_min,x2_max)
    
    plt.title(data_title)
    
    plt.grid(True)
    
def expandBorder(a, b): 
    d = (b - a) * 0.1
    return a - d, b + d
    
# 设定基本参数
row, col = 2, 4
linkages = ("ward", "complete", "average")
# 图形整体属性设定
cm = mpl.colors.ListedColormap(list('rgbmyc'))
plt.figure(figsize=(15, 9),facecolor='w')
# 圆形原始数据与子图的绘制
mydata_v(row, col, 1, data1, y1, 'Blob raw data')
for i, lk in enumerate(linkages):
    ac = AgglomerativeClustering(n_clusters=centers1, linkage=lk)
    y_predict = ac.fit_predict(data1)
    mdata_v(2, 4, i + 2, data1, y_predict, 'Blob ' + lk + ' data')
mydata_v(row, col, 1 + col, data2, y2, 'Moon raw data')
for i, lk in enumerate(linkages):
    ac = AgglomerativeClustering(n_clusters=centers2, linkage=lk)
    y_predict = ac.fit_predict(data2)
    mdata_v(2, 4, i + col + 2, data2, y_predict, 'Moon ' + lk + ' data')
plt.show()
```
