1.导入相关类库

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
```

2.获取鸢尾花数据

```python
iris = load_breast_cancer()
print '变量名字: ', iris.feature_names
print '标签名字: ', iris.target_names
print '样本个数: ', len(iris.data)
```

3.随机森林构造

```python
# 生成测试集和训练集
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1.0/4)
print '数据集样本数: {}, 训练集样本数: {}, 测试集样本数: {}'.format(len(X), len(X_train), len(X_test))
# 调用随机森林模型
clr = RandomForestClassifier()
# 训练模型
clr.fit(X_train, y_train)
# 训练的结果
print '训练结果: ', clr.score(X_train, y_train)
```

4.模型评估与可视化

```python
# 生成模型的预期值
lr_y_predict = clr.predict(X_test)
# 预测结果画图
x_test_len = range(len(X_test))
plt.figure(figsize=(16, 4), facecolor='w')
plt.plot(x_test_len, y_test, 'ro', markersize=4, zorder=3, label='real label')
plt.plot(x_test_len, lr_y_predict, 'go', markersize=10, zorder=2, label='predict label')
plt.legend(loc='lower right')
plt.xlabel(u'test data id', fontsize=18)
plt.ylabel(u'label', fontsize=18)
plt.title(u'Breast_cancer Classfication', fontsize=20)
plt.show()
```
